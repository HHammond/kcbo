```python
def generate_random_lognorm_data():
    import numpy as np
    import pandas as pd

    g1d = np.random.lognormal(mean=3, sigma=1, size=10000)
    g1l = ['A'] * g1d.shape[0]
    g2d = np.random.lognormal(mean=3, sigma=1, size=10000)
    g2l = ['B'] * g2d.shape[0]
    g3d = np.random.lognormal(mean=3.02, sigma=0.8, size=7000)
    g3l = ['C'] * g3d.shape[0]

    g1 = pd.DataFrame(data=g1d, columns=['value'])
    g1['group'] = g1l
    g2 = pd.DataFrame(data=g2d, columns=['value'])
    g2['group'] = g2l
    g3 = pd.DataFrame(data=g3d, columns=['value'])
    g3['group'] = g3l

    df = pd.concat([g1, g2, g3])

    return df

ln_data = generate_random_lognorm_data()
_ = kcbo.lognormal_comparison_test(ln_data, samples=400000)
_[1]
```
